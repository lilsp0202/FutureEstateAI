```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import skew 
import seaborn as sns

dframe= pd.read_csv("real_estate_train.csv")
data=dframe[['X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude','X6 longitude','Y house price of unit area']]

data.plot(kind='density', subplots=True, layout=(3,3), sharex=False,figsize=(20,10))


data_log=np.log(data)
data_log=data_log.replace([np.inf, -np.inf], np.nan)
data_log=data_log.dropna()

data_log.plot(kind='density', subplots=True, layout=(3,3), sharex=False,figsize=(20,10))



```

```python
log_x_plot=data_log[['X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude','X6 longitude']]
log_y_plot=data_log['Y house price of unit area']


for i in log_x_plot:
    plt.scatter(log_x_plot[i],log_y_plot)
    plt.xlabel(i)
    plt.ylabel('house price of unit area')
    plt.show()
```
